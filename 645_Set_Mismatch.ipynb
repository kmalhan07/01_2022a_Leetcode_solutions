### 645. Set Mismatch

### https://leetcode.com/problems/set-mismatch/

```python
def sm(nums):
    if 1 not in nums:
        return [[x for x in nums if nums.count(x) > 1][0], 1]
    elif len(nums) not in nums:
        return [[x for x in nums if nums.count(x) > 1][0], len(nums)]
    else:
        nums = sorted(nums)

        diff =  [nums[i + 1] - nums[i] for i in range(len(nums) - 1)]
        
        index_0 = [i for i, x in enumerate(diff) if x == 0][0]
        index_gt1 = [i for i, x in enumerate(diff) if x > 1][0]
        
        if index_0 < index_gt1:
            # This condition means a bigger number is missing
            return [index_0 + 1, index_gt1 + 1]
        if index_gt1 < index_0:
            # This condition means a smaller number is missing
            return [index_0 + 2, index_gt1 + 2]
```
