### 169. Majority Element

### https://leetcode.com/problems/majority-element/


**Constraints**:

$\bullet$ n == nums.length

$\bullet$ 1 <= n <= 5 * 10^4

$\bullet$ -10^9 <= nums[i] <= 10^9


```python
def majority_element(nums):
    
    n_min = len(nums) // 2
    
    nums_unique = set(nums)
    
    for element in nums_unique:
        if nums.count(element) > n_min:
            return element    
```

```python
nums = [2,2,1,1,1,2,2]
answer = majority_element(nums)
print("\nanswer = ", answer)
```

### Leetcode solution

```python
class Solution(object):
    def majorityElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return majority_element(nums)
    
    
def majority_element(nums):
    
    n_min = len(nums) // 2
    
    nums_unique = set(nums)
    
    for element in nums_unique:
        if nums.count(element) > n_min:
            return element     
```
