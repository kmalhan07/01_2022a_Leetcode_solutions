### 697. Degree of an Array

### https://leetcode.com/problems/degree-of-an-array/description/

```python
def degree(nums):
    
    # Find element with maximum occurences
    element_with_max_occurences = max(list(set(nums)), key = nums.count)
    n_max = nums.count(element_with_max_occurences)

    # But there could be more than one element with similar occurences
    element= []
    for x in set(nums):
        if nums.count(x) == n_max:
            element.append(x)
    
    # Find smallest length that contains the sub-array
    nums_r = nums[::-1]
    n_len = len(nums)
    degree = n_len
    for x in element:
        deg = (n_len - nums_r.index(x) - 1) - nums.index(x) + 1
        if deg <= degree:
            degree = deg
            
    return degree
```

```python
nums = [1,2,2,3,1]
answer = degree(nums)
print("\nanswer = ", answer)
```

### Leetcode

```python
class Solution(object):
    def findShortestSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        return degree(nums)
        
        
        
def degree(nums):
    
    # Find element with maximum occurences
    element_with_max_occurences = max(list(set(nums)), key = nums.count)
    n_max = nums.count(element_with_max_occurences)

    # But there could be more than one element with similar occurences
    element= []
    for x in set(nums):
        if nums.count(x) == n_max:
            element.append(x)
    
    # Find smallest length that contains the sub-array
    nums_r = nums[::-1]
    n_len = len(nums)
    degree = n_len
    for x in element:
        deg = (n_len - nums_r.index(x) - 1) - nums.index(x) + 1
        if deg <= degree:
            degree = deg
            
    return degree
```
